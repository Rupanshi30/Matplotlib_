## adding labels

```python
l1=[2,3,4]
l2=[4,5,6]
plt.plot(l1,l2)
plt.xlabel("x line")
plt.ylabel("y line")
plt.title("first plot");
#plt.show()
```

```python
#adding the markers and line size
l1=[2,3,4]
l2=[4,5,6]
plt.plot(l1,l2,marker="^",markersize=20)
plt.xlabel("x line")
plt.ylabel("y line")
plt.title("first plot");
#plt.show()
```

```python
# adding the formatting
# changing the color and linestyle
l1=[2,3,4]
l2=[4,5,6]
plt.plot(l1,l2,color="magenta",marker="^",linestyle="dashed",markersize=20)
plt.xlabel("x line")
plt.ylabel("y line")
plt.title("first plot");
#plt.show()
```

```python
# lineplot from array
x=np.arange(1,11)  #1-d array 
y=np.arange(12,22)
plt.plot(x,y,color="green",marker="^",linestyle="dashed",markersize=20)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line Plot using array");
```

```python
x=np.arange(1,11)  #1-d array 
y=x**3
plt.plot(x,y,color="green",marker="^",linestyle="dashed",markersize=20)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line Plot using array");
```

## Comparisons: Multiple lines in one figure

```python
x1=np.linspace(1,10,5)
x2=np.linspace(10,20,5)
plt.figure(figsize=(5,6))
x3=x2*2
x4=x2*3
plt.plot(x1,x2,label="2000") #labelling the lines
plt.plot(x1,x3,label="2005")
plt.plot(x1,x4,label="2010")
plt.legend()
plt.show()
```

```python
l=[1,2,3,4,5,6,7] ## product id
price=[400,5000,32002,60005,70089,90000,70064]  ##  total price
plt.plot(l,price,color="r",marker="*",markersize=8) #labelling the lines
plt.xlabel("Product Id")
plt.ylabel("Product-Price")
plt.title("Product price comparisons")
plt.show()
```

```python
l=[1,2,3,4,5,6,7] ## product id
plt.figure(figsize=(5,3),dpi=70)  ##like pixels
online_price=[400,500,322,65,789,90,764]  ##  online price
offline_price=[450,556,400,100,770,110,800] ### offline price
plt.plot(l,online_price,marker="*",label="zara") #labelling the lines
plt.plot(l,offline_price,marker="^",label="store")
plt.legend()
plt.xlabel("Product Id")
plt.ylabel("Product-Price")
plt.title("Product price comparisons")
plt.show()

## Insights: from this we can see online price of 1,2 and 3 products are low
```

```python
year=["2013","2014","2015","2016","2017","2018","2019","2020","2021","2022"]
online_sale=[5,8,8,9,10,12,8,7,20,22]
plt.plot(year,online_sale,marker="*",markersize=20)
plt.xticks(rotation=45)  ## changes in x-size rotation
plt.xlabel("year")
plt.ylabel("online_sale")
plt.title("Trend of the sales over the year")
plt.show()

#Insights: 2013-2020: ups and downs in the sale after 2020 drastic increase in sale
```

## Histogram 

- it is used for uni variate analysis of continuous type of data(mostly)
- It tells the frequency of the value of the data(value_counts tells frequency in pandas)


### histo eg: 1

```python
# data to display on plots 
x = [1, 2, 3, 4, 5, 6, 7, 4] 
# This will plot a simple histogram
plt.hist(x) ### bins= 10 default
# Title to the plot
plt.title("Histogram")
# Adding the legends
plt.legend(["bar"])
plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6, 7, 4] 
plt.hist(x,bins=5) ### bins= 10 default
plt.title("Histogram")
plt.legend(["bar"])
plt.show()
```

# Pie Chart()

- It is used for univariate analysis
- Mostly used for categorical
- It tells the distribution of categories of the data

```python
l=[200,300,310,800] ##answers in percentage
l1=["Python","D.A","D.S","ML"]
plt.pie(l,labels=l1);
```

```python
l=[200,300,310,800] 
l1=["Python","D.A","D.S","ML"]
plt.pie(l,labels=l1,autopct="%0.01f%%");  ## use autopct="%0.0f%%" for percent and 0.01 means only one number after decimal
```

```python
l=[200,300,310,800] ## analyse better in percentage
l1=["Python","D.A","D.S","ML"]
plt.pie(l,labels=l1,autopct="%0.0f%%"); ##no number after decimal
```

```python
l=[200,300,310,800] 
l11=["Python","D.A","D.S","ML"]
l1=[1,1,1,2]
plt.pie(l,labels=l11,explode=l1,autopct="%0.0f%%");  #formatting not necessary
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0,0,0,5] ## explode 1 means distance
plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0,0,0,1] ## explode 1 means distance
plt.pie(y, labels = mylabels, explode = myexplode,shadow=True)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mycolors=["lavender","blue","black","red"]
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0,0,0,1] ## explode 1 means distance
plt.pie(y, labels = mylabels,colors=mycolors,explode = myexplode,shadow=True)
plt.show() 
```
